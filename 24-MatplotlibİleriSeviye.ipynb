```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
numpyDizisi1 = np.linspace(0,10,20)
```

```python
numpyDizisi2 = numpyDizisi1 ** 3
```

```python
(benimFugur, benimEksen) = plt.subplots()
benimEksen.plot(numpyDizisi1,numpyDizisi2, color = "#3A95A8",alpha = 0.7) # Renk kodu ve saydamlık ayarı
benimEksen.plot(numpyDizisi2,numpyDizisi1, color = "#C96F23")
```

```python
(benimFugur, benimEksen) = plt.subplots()
benimEksen.plot(numpyDizisi1,numpyDizisi2, color = "blue", linewidth = 5.0) # Kalınlık ayarı için kullanılır.
benimEksen.plot(numpyDizisi2,numpyDizisi1, color = "yellow")
```

```python
(benimFugur, benimEksen) = plt.subplots()
benimEksen.plot(numpyDizisi1,numpyDizisi2, color = "blue", linestyle = "--") #Kesikli çizgi yapmak için kullanılır
benimEksen.plot(numpyDizisi2,numpyDizisi1, color = "green", linestyle = "-.")
benimEksen.plot(numpyDizisi2 + 100 ,numpyDizisi1 + 100, color = "red", linestyle = ":")
```
