**INSTANCE & ATTRIBUTE**

```python
class SuperKahraman():

  ozelGuc = "Gorunmezlik"

  def __init__(self,isimIpnut,yasInput,meslekInput):
    self.isimIpnut = isimIpnut
    self.yas = yasInput
    self.meslek = meslekInput
    print("Init çağrıldı")

  def ornekMethod(self):
    print(f"Ben süper kahramanım ve mesleğim: {self.meslek}")
```

```python
superman = SuperKahraman("Superman",30,"Gazeteci")
```

```python
superman.isim = "Clark Kent"
```

```python
superman.isim
```

```python
superman.ornekMethod()
```

```python
class Kopek():

  def __init__(self,yas=5):
    self.yas = yas

  def zaman(self):
    print(f"İnsanın yaşı: {self.yas*7}")
```

```python
benimKopek = Kopek()
```

```python
benimKopek.yas
```

```python
benimKopek.zaman()
```

**inheritance**

```python
class Hayvan():
  def __init__ (self):
    print("Hayvan sınıfı init çağrıldı")
  def method1(self):
      print("Hayvan sınıfı method1 çağrıldı")
  def method2(self):
      print("Hayvan sınıfı method2 çağrıldı")
```

```python
benimHayvan = Hayvan()
```

```python
benimHayvan.method1()
```

```python
benimHayvan.method2()
```

```python
class Kedi(Hayvan):
  def __init__(self):
    Hayvan.__init__(self)
    print("Kedi sınıfı çağrıldı")

  def miyavla(self):
    print("Kedi Miyavladı")
```

```python
benimKedi = Kedi()
```

```python
benimKedi.method1()
```

```python
benimKedi.miyavla()
```

Polymorphism

```python
class elma():
  def __init__(self,isim):
    self.isim = isim
  def bilgiVer(self):
    return self.isim + " 100 Kaloridir"
```

```python
class muz():
  def __init__(self,isim):
    self.isim = isim
  def bilgiVer(self):
    return self.isim + " 150 Kaloridir"
```

```python
elma =  elma("elma ")
```

```python
elma.bilgiVer()
```

```python
muz = muz("muz ")
```

```python
muz.bilgiVer()
```

```python
meyveListesi = [elma,muz]
```

```python
for meyve in meyveListesi:
  print(meyve.bilgiVer())
```
