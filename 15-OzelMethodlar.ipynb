```python
class meyveler():
  def __init__(self,isim,kalori):
    self.isim = isim
    self.kalori = kalori
  def __str__(self):
    return f"{self.isim} şu kadar kaloriye sahiptir: {self.kalori}"
  def __len__(self):
    return self.kalori
```

```python
muz = meyveler("Muz",150)
```

```python
muz.kalori
```
