```python
def bolmeIslemi(numara):
  return numara/2
```

```python
bolmeIslemi(20)
```

```python
benimListem = list(range(0,10))
```

```python
benimListem
```

```python
yeniListe = []
for eleman in benimListem:
  yeniListe.append(bolmeIslemi(eleman))
```

```python
yeniListe
```

**MAP**

```python
list(map(bolmeIslemi,benimListem))
```
