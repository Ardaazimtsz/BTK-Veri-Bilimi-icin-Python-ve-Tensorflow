**INPUT & RETURN**

```python
def merhabaDunya():
  print("Merhaba Dünya")
```

```python
merhabaDunya()
```

```python
def merhabaDunya(YazdirilacakIsim):
  print("Merhaba")
  print(YazdirilacakIsim)
```

```python
merhabaDunya("Arda")
```

```python
merhabaDunya("Python")
```

```python
def merhaba(isim = "Arda"):
  print("Merhaba")
  print(isim)
```

```python
merhaba()
```

```python
merhaba("Azim")
```

**ARGS & KWARGS**

```python
def yeniToplama(*args):
  return sum(args)
```

```python
yeniToplama(10,20,30,40,50)
```

```python
yeniToplama(10,20)
```

```python
def benimFonksiyonum(*args):
  print(args)
```

```python
benimFonksiyonum(10,20,30)
```

```python
def benimFonksiyonum(*args):
  return(args)
```

```python
benimFonksiyonum(10,20,30)
```

```python
def ornekFonksiyon(**kwargs):
  print(kwargs)
```

```python
ornekFonksiyon(muz=100, elma=200, ananas=300)
```
