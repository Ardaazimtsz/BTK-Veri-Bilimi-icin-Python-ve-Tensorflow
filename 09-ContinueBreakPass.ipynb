```python
benimListem = [5,10,15,20,25,30]
```

```python
for numara in benimListem:
  print(numara)
```

```python
for numara in benimListem:
  if numara ==  15:
    break
  print(numara)
```

```python
for numara in benimListem:
  if numara ==  15:
    continue
  print(numara)
```

```python
for numara in benimListem:
  if numara ==  15:
    pass
  print(numara)
```
