**Groupby**

```python
maasSozlugu = {"Departmanlar":["Yazılım","Hukuk","Yazılım","Hukuk","İnsan Kaynakları","Finans"],
               "Calışan İsmi":["Ahmet","Mehmet","Arda","Azim","Berke","Enes"],
               "Maas": [100,200,300,400,500,600],}
```

```python
maasDataFrame = pd.DataFrame(maasSozlugu)
```

```python
maasDataFrame
```

```python
grupObjesi = maasDataFrame.groupby("Departmanlar")
```

```python
grupObjesi.count()
```

```python
grupObjesi.mean()
```

```python
grupObjesi.min()
```

```python
grupObjesi.max()
```

```python
grupObjesi.describe()
```
