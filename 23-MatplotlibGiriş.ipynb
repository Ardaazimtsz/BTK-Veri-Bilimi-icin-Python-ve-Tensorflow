```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
numpyYasListesi = np.array([20, 25, 30, 35, 20, 45, 50, 55, 30])
```

```python
numpyKiloListesi = np.array([60, 65, 50, 75, 80, 85, 60, 15, 100])
```

```python
numpyYasListesi
```

```python
numpyKiloListesi
```

```python
plt.plot(numpyYasListesi,numpyKiloListesi,"y")
plt.xlabel("Yaş")
plt.ylabel("Kilo")
plt.title("Kilonun Yaşa Göre Değişim Grafiği")
plt.show() #python
```

```python
plt.plot(numpyKiloListesi,numpyYasListesi,"r")
```

```python
numpyDizisi = np.linspace(0,10,20)
```

```python
numpyDizisi
```

```python
numpyDizisi2 = numpyDizisi ** 3
```

```python
numpyDizisi2
```

```python
plt.plot(numpyDizisi,numpyDizisi2, "g")
```

```python
plt.plot(numpyDizisi,numpyDizisi2, "g--")
```

```python
plt.plot(numpyDizisi,numpyDizisi2, "b*-")
```

```python
plt.subplot(1,2,1)
plt.plot(numpyDizisi,numpyDizisi2, "r*-")
plt.subplot(1,2,2) #1 sıra olacak, 2kolon olacak,şuan 2.grafiği çiziyorum
plt.plot(numpyDizisi2,numpyDizisi, "g--")
```

```python
benimFigure = plt.figure()
figureAxes = benimFigure.add_axes([0.8,0.8,0.3,0.3])
figureAxes.plot(numpyDizisi,numpyDizisi2,"g")
figureAxes.set_xlabel("X Ekseni Veri İsmi")
figureAxes.set_ylabel("Y Ekseni Veri İsmi")
figureAxes.set_title("Grafik Başlığı")
```

```python
figure2 = plt.figure()
eksen1 = figure2.add_axes([0.1,0.1,0.9,0.9]) # Başlangıç koordinatlarını ve boyutunu ayarlar
eksen2 = figure2.add_axes([0.2,0.5,0.3,0.3])

eksen1.plot(numpyDizisi,numpyDizisi2,"g") # Verileri ve rengi belirleriz
eksen1.set_xlabel("X Ekseni") # X eksenine isim verir
eksen1.set_ylabel("Y Ekseni") # Y eksenine isim verir
eksen1.set_title("Ana Grafik Başlık") # Grafiğin ana ismini verir

eksen2.plot(numpyDizisi2,numpyDizisi,"b")
eksen2.set_xlabel("X Ekseni")
eksen2.set_ylabel("Y Ekseni")
eksen2.set_title("Küçük Grafik Başlık")
```

```python
yeniFigur = plt.figure()

yeniEksen = yeniFigur.add_axes([0.1,0.1,0.9,0.9])
yeniEksen.plot(numpyDizisi,numpyDizisi**2, label ="numpyDizisi ** 2" )
yeniEksen.plot (numpyDizisi,numpyDizisi**3, label = "numpyDizisi ** 3")
yeniEksen.legend(loc = 2) # Label yer değiştirmek için kullanılır.
```

```python
yeniFigur = plt.figure(figsize = (9,9)) # Boyut ayarlamak için kullanılır.

yeniEksen = yeniFigur.add_axes([0.1,0.1,0.9,0.9])
yeniEksen.plot(numpyDizisi,numpyDizisi**2, label ="numpyDizisi ** 2" )
yeniEksen.plot (numpyDizisi,numpyDizisi**3, label = "numpyDizisi ** 3")
yeniEksen.legend(loc = 2) # Label yer değiştirmek için kullanılır.
```

```python
yeniFigur = plt.figure(dpi = 1000) # Çözünürlük ayarlamak için kullanılır.

yeniEksen = yeniFigur.add_axes([0.1,0.1,0.9,0.9])
yeniEksen.plot(numpyDizisi,numpyDizisi**2, label ="numpyDizisi ** 2" )
yeniEksen.plot (numpyDizisi,numpyDizisi**3, label = "numpyDizisi ** 3")
yeniEksen.legend(loc = 2) # Label yer değiştirmek için kullanılır.
```

```python
yeniFigur.savefig("benimfigür.png", dpi=200) #Kaydetmek için kullanılır
```
